```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
```

```python
from google.colab import files # Import the 'files' object from the 'google.colab' module
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
import numpy as np # Import numpy and alias it as 'np'
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array

img = load_img(filename,target_size=(224,224))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array,axis=0) # Now 'np' is defined and can be used
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.5,1.5]
)
```

```python

import matplotlib.pyplot as plt # Import matplotlib.pyplot and alias it as 'plt'

i = 0
for batch in datagen.flow(img_array,batch_size=1):
  plt.figure(figsize=(2,2)) # Now 'plt' is defined and can be used
  plt.imshow(batch[0].astype('uint8'))
  plt.axis('off')
  i+=1
  if i >5:
    break
```
